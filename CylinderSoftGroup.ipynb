```python
import os
# 00 , 01,  08, 07
# Specify the path to the folder you want to count files in
folder_path = '/content/drive/MyDrive/Repositories/SoftGroup/dataset/kitti/dataset/sequences/00/voxels'

# Initialize a counter for files
file_count = 0

# Check if the specified path is a directory
if os.path.isdir(folder_path):
    # List all files in the directory
    files = os.listdir(folder_path)

    # Iterate through the files and count them
    for file in files:
        if os.path.isfile(os.path.join(folder_path, file)):
            file_count += 1

    # Display the number of files
    print(f'The folder "{folder_path}" contains {file_count} files.')
else:
    print(f'The path "{folder_path}" is not a valid directory.')
```

```python
import os
import hashlib

# Specify the path to the folder containing the files
folder_path = '/content/drive/MyDrive/Repositories/SoftGroup/dataset/kitti/dataset/sequences/00/labels'

# Dictionary to store file hashes and their corresponding paths
file_hashes = {}

# Function to calculate the hash of a file
def calculate_file_hash(file_path):
    hasher = hashlib.sha256()
    with open(file_path, 'rb') as file:
        while True:
            data = file.read(65536)  # Read in 64k chunks
            if not data:
                break
            hasher.update(data)
    return hasher.hexdigest()

# Iterate through all files in the folder
for root, dirs, files in os.walk(folder_path):
    for file_name in files:
        file_path = os.path.join(root, file_name)
        file_hash = calculate_file_hash(file_path)

        # Check if the hash already exists in the dictionary
        if file_hash in file_hashes:
            # If a duplicate is found, delete the duplicate file
            print(f"Removing duplicate: {file_path}")
            os.remove(file_path)
        else:
            # If not a duplicate, store the hash and file path in the dictionary
            file_hashes[file_hash] = file_path

print("Duplicate files removed.")
```
